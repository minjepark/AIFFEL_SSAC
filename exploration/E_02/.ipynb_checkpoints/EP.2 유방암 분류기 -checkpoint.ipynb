```python
from sklearn.datasets import load_breast_cancer 
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

## (2)  

```python
breast = load_breast_cancer()
print(breast.keys())
data = breast.data
label = breast.target
```

## (3)  

```python
print(breast.feature_names)
(data.shape)
print(len(breast.target_names))
print(breast.target_names)
```

```python
print(breast.DESCR)
```

```python
#        . 
import pandas as pd
alldata = pd.DataFrame(data= data, columns=breast.feature_names)
alldata
```

```python
alldata['label'] = breast.target
alldata
```

```python
#   .  
import seaborn as sns
sns.boxplot(alldata['label'],alldata['mean radius'])
```

## (4)train,test   

```python
x_train, x_test, y_train, y_test = train_test_split(data, label,random_state=42, test_size=0.2)
```

## (5)   
### 1 Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
print(classification_report(y_test, y_pred))
```

### 2 Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
print(classification_report(y_test,y_pred))
```

### 3 SVM 

```python
from sklearn import svm
svm = svm.SVC()
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)
print(classification_report(y_test,y_pred))
```

### 4  SGD Classifier

```python
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier()
sgd.fit(x_train,y_train)
y_pred = sgd.predict(x_test)
print(classification_report(y_test,y_pred))
```

### 5 Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(loss='log,max_iter=10, random_state)
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print(classification_report(y_test,y_pred))
```
