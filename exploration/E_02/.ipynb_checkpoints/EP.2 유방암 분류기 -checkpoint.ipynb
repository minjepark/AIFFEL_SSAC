# 유방암 데이터 


+ 총데이터: 569개 
+ feature:39개  
+ class: 2개 (malignant,benign)

data는 569명의 조직들의 위치, 거리등을 수치적으로 나타내어 종양이 있는지 없는지를 뱉어준다. 

```python
from sklearn.datasets import load_breast_cancer 
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

## (2)데이터 준비 

```python
breast = load_breast_cancer()
print(breast.keys())
data = breast.data
label = breast.target
```

## (3)데이터 이해하기 

```python
print(breast.feature_names)
(data.shape)
print(len(breast.target_names))
print(breast.target_names)
```

```python
print(breast.DESCR)
```

```python
#데이터가 수치적인데 어떻게 생긴건지 와닿지가 않아서 데이터 프레임으로 변환시켜보았습니다. 
import pandas as pd
alldata = pd.DataFrame(data= data, columns=breast.feature_names)
alldata
```

```python
alldata['label'] = breast.target
alldata
```

```python
#데이터의 분포를 확인해 보겠습니다.  
import seaborn as sns
sns.boxplot(alldata['label'],alldata['mean radius'])
```

## (4)train,test 데이터 분리 

```python
x_train, x_test, y_train, y_test = train_test_split(data, label,random_state=42, test_size=0.2)
```

## (5)다양한 모델로 학습시켜보기 
### 1번 Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
print(classification_report(y_test, y_pred))
```

### 2번 Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
print(classification_report(y_test,y_pred))
```

### 3번 SVM 

```python
from sklearn import svm
svm = svm.SVC()
svm.fit(x_train, y_train)
y_pred = svm.predict(x_test)
print(classification_report(y_test,y_pred))
```

### 4번  SGD Classifier

```python
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier()
sgd.fit(x_train,y_train)
y_pred = sgd.predict(x_test)
print(classification_report(y_test,y_pred))
```

### 5번 Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(loss='log,max_iter=10, random_state)
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print(classification_report(y_test,y_pred))
```

## (6)모델을 평가해 보기

걸렸는데 못 걸러 낸다면 정말 큰 문제이다.   
그래서 양성을 놓치지 않는 recall이 높은 random forest모델을 사용하겠습니다.
