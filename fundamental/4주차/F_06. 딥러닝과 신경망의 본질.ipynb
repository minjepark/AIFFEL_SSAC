## 6-2 Deep learning 이란?

인공지능: 인간의 설계에 의한 분류가 아닌 스스로 기준을 제시하는 것.       
머신러닝: 데이터를 제공하면 데이터를 분석하고, 패턴을 학습하며 학습한 것을 토대로 예측을 한다.     
딥러닝: 스스로 데이터의 feature를 찾는다.  날 것의 데이터를 주면 그 안에서 내재된 표현을 추출한다는 것! 그것을 잘~~추출하는게 목표이다.


### 왜 deeplearning인가?

+ 합성된 함수를 학습시켜서 풍부하면서도 유용한 '내재적 표현'을 찾아내는 machine을 구축하는 것이 목표이다.

### 표현이란 무엇일까?

+ 세상 모든 것은 '데이터'라고 할 수 있습니다. 데이터는 다양하게 '표현'될 수 있다고 합니다.  
+ 표현 (representation)

  

하나의 데이터를 분자, 이미지, 표이 하나의 데이터를 여러가지로 표현이 가능합니다.   

데이터의 내재된 표현이란, 데이터가 가지고 있는 정보에서 추상적인 것을 추출한다는 것인데 추상적이라는 것은 인간의 개입이 많은 그런 데이터표현법이다.   데이터만 주고 어떻게 표현을 추출하는냐?가 문제인데 그거는 gradient descent로 알 수 있습니다. 


### 딥러닝을 관통하는 철학 (1) 행동주의

강화학습을 설명할 때 스키너상자를 예로 든다. 쥐가 상자 안에 있고 레버를 누르면 음식을 먹는다. 모르고 있다가 계속해서 레버를 누르면 밥을 먹게된다는 걸 알게되면 그것을 여러번한다(그 행동이 강화된다)

### 딥러닝을 관통하는 철학 (2) 인지주의   

인간은 비둘기와 달리 정보를 처리하는 '의식'이 있다. 
인지주의:  자극 > (정보처리) > 행동 으로 행동주의보다 onestep이 늘었다.   

이 정보처리 과정은 물론 **뇌**에서 진행합니다. 


### 딥러닝을 관통하는 철학 (3) 연결주의   

연결주의: 정보처리는 뇌에 있는 수많은 뉴런의 연결고리를 통해 이루어진다.   
    
인지주의를 넘어서 연결되어 있다는 연결주의가 뉴런을 설명하기 좋을 듯하다. 

처음은 백지이지만 하다보면 경험이 쌓이고 경험에 의해 학습해 간다는 의미이다.

### 6-9. 신경망의 본질(3) 함수의 역할 -상 

**함수의 역할 3가지**

+ relation: x와 y의 관계를 나타낼 수 있는 도구   
    + 'y는 x의 함수이다.'라는 의미는 y는 x에 종속적이라는 의미입니다.        
    
+ transformation: (x)를 변화해주는 도구 
    + linear transformation이 되려면 직선의 형태 & 변환후에도 원점은 이동하지 않아야 합니다.   
    + transformation은 옮기는 역할을 합니다.    
    
 + x의 공간에서 y의 공간으로 매핑시켜주는 도구 

```python
import numpy as np

def linear_transformation(X,A):
    transformed_vector = np.matmul(X,A)
    return transformed_vector
```

```python
x = np.array([-1,2])
a = np.array([[1,-2],
              [3,0]])

linear_transformation(x,a)
```

## 6-11. 신경망의 본질 (5) 함수와 모델 

함수는 형태가 고정되어 있는 경우가 대부분입니다. 방정식의 해가 정해져 있잖아요.   
하지만 우리가 찾는 모델은 함수의 형태, 정해진 해도 없고 그냥 **데이터**만 가지고 있습니다.    

그렇기에 머신러닝/딥러닝은 정확한 해를 찾는게 아니라 **그나마 가장 잘 근사할 수 있는 함수**를 찾는 것입니다.  



### 그나마 잘 맞는 함수를 찾기 위해   

1) 먼저 모델을 어떤 함수로 나타낼 것인지를 정해야 합니다.  
    + 비정형데이터는 필연적인 신경망 모델을 사용할 듯   
    + inductive bias or Prior라는 이야기는 **데이터를 설명할 수 있는 최적의 함수가 특정한 함수 공간에 존재할 것이라는 가설을 의미합니다.**       
    + 모델이 세상 모든 데이터를 보지 못하니 한번도 본 적없는 데이터에 대한 예측도 잘 하기를 (generalization) 바랍니다.  
    + 어떤 문제를 풀고자 할 떄 알맞는 함수 형태로 모델을 선택하는 것은 데이터에 대한 도메인 지식이 있는 사람의 몫이고, 더 좋은 모델을 찾기 위해서는 
     그 데이터에 알맞는 함수공간 inductive bias를 잘 설정하는 것이 중요합니다.   
     
2) 함수의 형태를 정했다면, 해당 함수 공간 안에서 최적의 함수를 찾아나갑니다.  

    +이 과정은 보통 모델을 학습시킨다고 이야기 하는 것이다. 엉뚱한 값을 내던거를 적절한 값을 출력하도록 '보정'해 나가는 과정이다.
