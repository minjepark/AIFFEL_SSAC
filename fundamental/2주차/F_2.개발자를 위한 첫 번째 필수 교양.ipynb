### 2-5. 코드의 모든 발자취를 기록할 수 있을까? (2) GIT와 GitHub  

**GIT** 
- 개발을 진행하며 작성하는 소스코드가 업데이트 되는 버전을 기록해 두고 관리할 수 있는 소스코드 관리 시스템  
- 버전 기록을 저장
- 리눅스 창시자가 만든 오픈소스 툴

**GitHub**    
- Git으로 관리하는 프로젝트를 호스팅하고, 시간과 공간의 제약없이 협업할 수 있는 온라인 시스템  
- git으로 기록한 버전을 다른 사람과 공유하며 헙업할 수 있도록 하는 시스템 
- git을 더 쉽게 사용할 수 있게 하는 서비스

 
 

### 2-7 GitHub에 잔디심기 (1)로컬저장소  

1. 로컬의 Git에 GitHub의 계정 정보 등록하기 

터미널의 열고 아래의 명령어를 입력하세요. 

`git config --global user.email "my_email@gmail.com"`  

`git config --global user.name "my_username"`

`my-email@gmail.com` 과 `my-username`부분은 자신의 email주소와 username으로 변경해서 입력하면 된다고 합니다. 

### 내 컴퓨터에 로컬 저장소 만들기

`$cd ~`   
`$mkdir directoryname`   
directoryname에 원하는 이름으로 저장해주세요.  
`$ cd directoryname` 으로 가셔서   
`$ git init` init는 initialization의 약자로 이제부터 git이 지금있는 폴더에서 발생하는 변화를 기록해 둔다라는 의미입니다. 
`$ ls -a` a는 all로 ls만 써서 보이지 않는 모든 것이 확인 가능합니다.


`$ cd .git`   .이 있으면 숨겨진 파일이라는 의미입니다. 

`$ ls`를 눌러서 확인하면 끝!!ㅎ



### 2-8 GitHub 첫 번째 잔디심기, 어렵지 않아요! (2)변화추적하기 

+ git이 있는 폴더에서 `echo '# 리드미에 쓸 내용' >>  README.md` 로 하면 리드미 파일이 생성됨  

+ `cat README.md` `cat`명령어로 작성된 내용을 확인 할 수 있습니다. 

+ `git status' :변화를 감지하겠습니다.   

+ `git add 파일명 파일명` : 파일을 올리기위해 준비하겠습니다.   
+ `git add .` 변화된 것을 모두 올리겠다. ㅎ


+ `git commit -m '설명'` : 파일을 올립니다!!!! 


## 막간을 이용한 git!  

git의 장점은 대부분의 명령을 로컬에서 실행한다는 점입니다. 

git 생성하기   
로컬 저장소로 사용할 폴더를 생성하여 해당 폴더로 이동합니다.  
`git init` 깃 폴더가 생성며 프로젝트 관리를 위한 파일들과 해당 프로젝트에만 적용할 config 파일 등이 들어 있습니다.   
`git status` 현재 저장소 내 파일들의 상태를 확인해 볼 수 있습니다.  
`git add` 

git은 작업폴더 add > 인덱스 commit > 저장소

---

`git remote -v` 현재 연결되어있는 원격저장소 상태를 알려준다. 
`git remote rm main origin` 연결된 저장소를 삭제 가능합니다.
