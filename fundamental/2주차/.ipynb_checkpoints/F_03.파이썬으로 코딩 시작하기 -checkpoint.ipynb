## 3-9. 제어문 

+ if 명제 : if키워드 다음의 명제가 참일 때만 실행됩니다. 
+ 비교연산자: <,>,<=,>=,==,!= (같다를 뒤에 쓴다)
+ 논리연산자: and(모두 참) , or(둘 중 하나) 
+ ` if (number >0) and (number%2==0):`이렇게 식과 식은 괄호를 써줘야한다. 


```python
def print_whether_positive_or_negative(number):
    if (number >0) or (number%2!=0):
        print(number)
    elif number == 0:
        print('0')
    else:
        print('-')
    
```

```python
print_whether_positive_or_negative(5)
print_whether_positive_or_negative(3)
print_whether_positive_or_negative(-1)
print_whether_positive_or_negative(0)
```

## 3-11. 제어문(2)'while': 조건이 참인 동안 반복

+ 무한루프에 빠지지 않도록 유의! 

```python
def fibonacci(n):
    if n<=2:
        return 1 
    else:
        return fibonacci(n-2) + fibonacci(n-1)

n = 1 

while n <= 20:
    print(fibonacci(n))
    n = n+1
print('끝')
```

## 3-15. 자료형(2)부동소수점 floating_point number, 'float'

+ 정수를 제외한 실수를 부동소수점 
+ 정수 1을 소수로 만들고 싶다면 `float(1)`이라고 하면 됩니다. 
+ 소수점의 표현방법: 컴퓨터는 0과 1로만 표현되기 때문에 소수점 아래의 소수는 가장 근사치의 값으로 저장됩니다.   
    + 고정소수점: 정수를 표현하는 비트 수와 소수를 표현하는 비트 수를 미리 정해 놓고, 해당 비트만큼만 사용해서 숫자를 표현하는 방식
    + 부동소수점: 소수의 위치를 고정하지 않고 그 위치를 나타내는 수를 따로 적는 것으로, 유효 숫자를 나타내는 가수와 소수점의 위치를 풀이하는 지수로 나타낸다.
