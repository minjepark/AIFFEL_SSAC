### 운영체제란?

### 1-3. 터미널을 자유자재로, 내 친구 삼기  

+ `whoami` :이름 찾기  
+ `pwd` : 현재 경로 출력  
+ `cd` : change directory  
+ `cd ..` : 상위폴더로 이동  
+ `cd ~` : 홈폴더로 이동 `cd`만 쳐도 이동 가능
+ `exit`: 창을 닫는다.   
+ `ls` : 현재 디렉토리 목록  
+ `mv 이동시킬 폴더 이동할 위치 폴더 ` : 파일을 이동합니다. 
+ `rm 파일명`  : 파일 지우기
+ `rm -r 폴더명` : 폴더 지우기   `-p`를 쓰면 파일이 지워진다. 
+ `cp -r '복사할 폴더' '이동할 위치경로'`: 파일 복사 



만약에 폴더를 동일선상에 있는 폴더가 아니라면 이동시킬 목적지 폴더에 가서   
`pwd`로 해당경로를 확인 후 복사하여 `rm 이동할 폴더 목적지 전체경로` 이렇게 써야한다.

### 1-4  Terminal, 또 무엇을 할 수 있을까?  

터미널의 역할:**패키지 설치 및 관리**를 할 수 있다.  
패키지란: 라이브러리보다 큰 개념이 패키지이지만 같이 혼용해서 사용한다.  
**우분투에서 패키지 관리를 위해 사용하는 apt-get**  
root계정이 아니라는 가정하에 사용할게요~! root가 있다면 sudo를 빼고 사용하셔도 됩니다!  



`apt list --installed`: 설치된 패키지 리스트 확인  
`apt list --installed | gep (패키지명)`: 내가 필요한 패키지가 설치되어 있는지 확인이 가능 

`sudo apt-get update`:  패키지에 대해 새롭게 업데이트 된 버전이 있따면 전부 업그레이드를 하라는 명령어입니다.                           
                       업데이트는 기존버전과 충돌이 있을 수있음
                       
`apt-get install -y cmatrix`: -y는 중간에 나오는  y/n를 yes로 하겠다는 의미입니다. 


`install` : 새로운 패키지를 설치   
`sudo apt-get install 패키지명`  

`remove`: 설치된 패키지를 제거 ** 패키지 관련 데이터는 삭제되지 않음 **         
`sudo apt-get remove 패키지명`  

`purge`: 패키지와 관련 데이터를 **모두** 삭제  
`sudo apt-get purge 패키지명`  

`--reinstall install`: 패키지를 재설치, 설치된 패키지에 문제가 있을 때 사용하면 유용  
`sudo apt-get --reinstall install`  

`clean`: 패키지 설치를 위해 다운받은 데이터 삭제   
`sudo apt-get clearn`  

`autoremove`: 필요없는 혹은 사용하지 않고 있는 패키지를 자동으로 삭제.     
`sudo apt-get autoremove`  

`update`: 패키지 저장소에서 패키지 리스트를 새로 다운받아 업데이트  
`sudo apt-get update`  

`upgrage`: 설치된 모든 패키지를 상위버전으로 업그레이드   
`sudo apt-get upgrade`  

`dist-upgrade`: 설치된 모든 패키지의 의존성을 검사하여 업그레이드, 만약 업그레이드 중 새로 추가된 의존성 패키지가 있따면 추가설치 진행.  
`sudo apt-get dist-upgrade`

`source`: 패키지를 설치하지는 않고 소스파일만 다운받음   
`sudo apt-get source 패키지명`  

`build-dep`: source를 통해 다운받은 패키지를 의존성을 검사하여 설치. 만약 설치도중 의존성 패키지가 있다면 추가 설치 진행.  
`sudo apt-get build-dep 패키지명`  

`apt-cache show`: 패키지의 ㅡ상세한 정보를 출력  
`sudo apt-cache show 패키지명`  

`apt-cache search`: 패키지명을 검색  
`sudo apt-cache search 패키지명`  

패키지명을 입력할 때 Ubuntu18.04 LTS, bash 기준으로 패키지명이 어느 정도 기억나는데 뒷부분이 기억나지 않을 때는 Tab키를 연속으로 2번 눌러주면 됨

### 1-5. 가상환경으로 패키지를 조금 더 유연하게 관리해보자 

가상환경이란? 

+ 독립된 공간 마련
+ 프로젝트에서 다른 버전을 써야 한다면 가상환경을 제공 
+ 파이썬버전도 다르게 넣을 수 있습니다.  
+ 리눅스에서 `python3 -m venv '가상환경명'`  
`cd /'가상환경명'`    

`source bin/activate (가상환경명) ~/가상환경명` 먼가 이상하지만..음 그렇다니깐 일단! 

https://dojang.io/mod/page/view.php?id=2470

**패키지**: 특정 기능을 위해 여러 함수 또는 클래스를 담고 있는 보따리  
**가상환경**: 컴퓨터에 설치된 패키지 간의 충돌 또는 버전관리를 위해 가상으로 나누어서 사용하는 환경    
**아나콘다**: 가상환경을 관리하는 툴   
**환경변수**: 콘다의 명령어를 터미널에서 사용할 수 있도록 만들어주는 것


아나콘다 설치   

파일 설치 > 환경변수 터미널과 연결 >
설치여부를 확인합니다. 리눅스환경에서는 `which conda`, 윈도우에서는 `where conda`  

`bash 프로그램` :설치가 됨 



`source ~/.bashrc` conda를 터미널에서 쓸 수있도록 하는 명령어 

가상환경 생성 및 실행    

`conda create -n my_env_name python=3.7`

`pip list`: 설치파일 전체보기   
`pip list | grep tensorflow`: 특정패키지 설치여부 확인   
`conda deactivate` 가상환경 비활성화시키기  
`conda env remove -n '가상환경명'`
