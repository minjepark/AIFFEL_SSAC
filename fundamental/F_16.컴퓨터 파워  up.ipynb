## multitasking이란? 

1) 컴퓨터 자원을 최적화 하는 방법으로 2가지 방법이 있다.
    1. 병렬처리 컴퓨터 2.분산처리 컴퓨터
   
### 동시성
    + 한명이 여러일 하기 
    + 하나의 processor가 여러가지 일을 동시에 수행하는 것을 말합니다. 
    + 특정순간에는 한가지만 수행하겠지만  task전환을 통해 효율적으로 다른 일을 처리합니다. 

### 병렬성
    + 여러명이 여러일 하기 
    + 유사한 task를 여러 processor가 진행합니다. 
    + processor의 갯수만큼 빨라집니다. 
    
동시성과 병렬성이 동시에 적용하게 된다면 더 빨리 일을 처리할 수 있습니다.      
병렬성의 효율을 극대화하는 것은 동시성을 극대화시키는 방법입니다. 


## 동기 vs 비동기?

어떤 일을 바로하지 못하고 대기해야하는 일을 '바운드 되었다'고 한다. 
바운드시에 계속 기다릴지 아니면 종료할지를 고민합니다. 

### 동기
    + 앞작업이 끝나기를 기다립니다. (순서대로 진행함)

### 비동기
    + 기다리면서 다른 일을 처리하는게 비동기입니다.   
    예를들어 라면 물이 끓기를 기다리고 면을 넣는다면 동기, 라면 물이 끓기를 기다리면서 언니랑 이야기를 한다면 비동기입니다. 
    
## I/O bound vs CPU bound

컴퓨터가 일을 수행하면서 속도제한이 걸린다면 이 두가지에 해당합니다. 

+ **I/O bound**: 입력과 출력에서 데이터 처리에 시간이 소요될 때.
+ **CPU bound**: 복잡한 숫자 수식계산이나 그래픽 작업과 같이 엄청난 계산이 필요할 때.

## thread, process,profiling

컴퓨터 실행시 컴퓨터에서 일어나는 일? 

### process
    + 프로그램을 구동하여 프로그램 자체와 프로그램의 상태가 메모리상에서 실행되는 작업단위.
    + 전용 메모리 공간을 갖습니다. 
    
### thread
    + program 내에서 엄밀히 따지만 프로세스 내에서 실행되는 흐름의 단위같은 작업을 좀 더 빠르게 처리하기 위해서 여러개의 thread를 생성합니다. 
    + 해당 프로세스 내의 전용공간을 활용합니다. 
    
### profiling
    + 코드의 병목을 찾아내고 성능을 측정해주는 도구입니다. 
    + 코드에서 시스템의 어느 부분이 느린지 혹은 어디에 ram이 많이 사용하고 있는지를 확인하고 싶을 때 사용합니다. 
    + windows 작업관리자를 통해 프로그램의 상태를 확인하는 것을 말합니다. 
    + 어플리케이션에서 가장 자원이 집중되는 지점을 찾아내는 것입니다. 
    
## sacle up vs scale out

### scale up
    + 한대의 컴퓨터 성능을 최적화 하는 것입니다.
    + 컴퓨터의 성능을 확장합니다. 
    
### scale out 
    + 여러대의 컴퓨터가 한대처럼 사용합니다.
